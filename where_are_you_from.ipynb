```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.lines import Line2D
```

```python
W = -0.2
St = 1
R = 1
t0 = 0
tf = 30
a = 1

load_data = np.load(f'hills_i_W_{W}_St{St}_R{R}_tf{tf}.npy')
initial_dist = np.sqrt(np.square(load_data[:,0]) + np.square(load_data[:,1]))
final_dist = np.sqrt(np.square(load_data[:,2]) + np.square(load_data[:,3]))
inside = np.column_stack((initial_dist < a, final_dist < a))
entrained_particles_index = np.where((inside[:,0] == False) & (inside[:,1] == True))[0]
detrained_particles_index = np.where((inside[:,0] == True) & (inside[:,1] == False))[0]

entrained_particles = load_data[entrained_particles_index, :2]
detrained_particles = load_data[detrained_particles_index, :2]

fig, ax = plt.subplots()
ax.scatter(entrained_particles[:,0], entrained_particles[:,1], c='red',s=0.1, label='entrained')
ax.scatter(detrained_particles[:,0], detrained_particles[:,1], c='blue',s=0.1, label='detrained')
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_aspect('equal')
ax.legend()

# Plot a dashed circle
circle_0 = Circle((0,0), 1, fill=False, color='black', linestyle='--', zorder=5)
ax.add_patch(circle_0)

```
